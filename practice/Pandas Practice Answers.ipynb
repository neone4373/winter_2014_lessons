Write several functions to apply to the data frame that attempt to organize the data set using strings instead of floats, like our short_or_long function from lecture. Use this to best summarize your data.

```python
def short_or_long(x):
    """Not mentioned in class, sorry for the confusion"""
    if x > 4:
        return 'long'
    else:
        return 'short'
```

```python
def petal_width_cat(x):
    if x < 0.6:
        return "short"
    elif x < 1.6:
        return "medium"
    else:
        return "long"
    
def petal_length_cat(x):
    if x < 3:
        return "short"
    elif x < 4.8:
        return "medium"
    else:
        return "long"
```

CHALLENGE QUESTION: From everything above, you should be able to predict relatively accurately for each row what kind of flower it is (without using the Name column as an obvious hint). Write a function that uses the data with if and else statements to attempt to classify each row.

```python
def flower_type(df):
    for _index_ in df.index:
        row = df.ix[_index_]
        pw = petal_width_cat(row["PetalWidth"])
        pl = petal_length_cat(row["PetalLength"])
        name = row["Name"]
        if pw == "short" and pl == "short": 
            pred = "Iris-setosa"
        elif pw == "medium" and pl == "medium":
            pred = "Iris-versicolor"
        elif pw == "long" and pl == "long":
            pred = "Iris-virginica"
        else:
            pred = "???"
        print "Pedicted: %s, Actual: %s" % (pred, name)

flower_type(data)
```
